# Comparison between linear and nonlinear aerodynamics coupled with a nonlinear beam

## Non trim initial condition (and linearisation point)

u_inf = 20 m/s

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nonlin_data = pd.read_csv('nontrim_init/nonlinear_u20_nontrim.csv')
lin_data = pd.read_csv('nontrim_init/linear_u20_nontrim.csv')
```

```python
fig = plt.figure(figsize=(12,10))
dt = 1 / 6 / 20

plt.plot(nonlin_data['Time']*dt, nonlin_data['app_forces (2) (stats)'],
        ls='-',
        lw=2,
        color='k',
        label='Nonlinear UVLM')
plt.plot(lin_data['Time']*dt, lin_data['app_forces (2) (stats)'],
        ls='--',
        lw=2,
        color='k',
        label='Linear UVLM')

plt.rcParams.update({'font.size': 16})

plt.legend()
plt.grid()
plt.xlabel('Time, t [s]')
plt.ylabel('$F_z$ [N]')
plt.title('Non-trim initial condition')
plt.xlim([0,5])

ax = fig.add_axes([0.2, 0.16, 0.25, 0.25])
ax.plot(nonlin_data['Time']*dt, nonlin_data['app_forces (2) (stats)'],
        ls='-',
        lw=2,
        color='k',
        label='Nonlinear UVLM')
ax.plot(lin_data['Time']*dt, lin_data['app_forces (2) (stats)'],
        ls='--',
        lw=2,
        color='k',
        label='Linear UVLM')
ax.set_xlim([0,0.5])
ax.set_ylim([10,30])
ax.grid()

plt.savefig('/home/ng213/Documents/PhD/WeeklyMeetings/Wk10_181128/Nontrim_f_steady.eps')
```

```python
fig = plt.figure(figsize=(12,10))
dt = 1 / 6 / 20

plt.plot(nonlin_data['Time']*dt, nonlin_data['app_moments (1) (stats)'],
        ls='-',
        lw=2,
        color='k',
        label='Nonlinear UVLM')
plt.plot(lin_data['Time']*dt, lin_data['app_moments (1) (stats)'],
        ls='--',
        lw=2,
        color='k',
        label='Linear UVLM')

plt.rcParams.update({'font.size': 16})

plt.legend()
plt.grid()
plt.xlabel('Time, t [s]')
plt.ylabel('$M_y$ [N m]')
plt.title('Non-trim initial condition')
plt.xlim([0,5])

ax = fig.add_axes([0.2, 0.16, 0.25, 0.25])
ax.plot(nonlin_data['Time']*dt, nonlin_data['app_moments (1) (stats)'],
        ls='-',
        lw=2,
        color='k',
        label='Nonlinear UVLM')
ax.plot(lin_data['Time']*dt, lin_data['app_moments (1) (stats)'],
        ls='--',
        lw=2,
        color='k',
        label='Linear UVLM')
ax.set_xlim([0,0.5])
ax.set_ylim([0,5])
ax.grid()
plt.savefig('/home/ng213/Documents/PhD/WeeklyMeetings/Wk10_181128/Nontrim_m_steady.eps')
```

Comparison in processing times:
* Nonlinear time to solve 1680s
* Linear time to solve 1890s

_However_ the linear solution really took a lot less time (parallelizing?)
