# Goland Wing Linear vs Nonlinear UVLM comparison

# Straight Wing

u_inf = 50

gust_intensity = 0.001

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size':16})
main_chord = 1.8288
```

```python
lin = pd.read_csv('goland_straight_lin.csv')
nlin = pd.read_csv('goland_straight_nlin.csv')

dt = main_chord / 6/ 50
```

```python
fig = plt.figure(figsize=(12,10))

plt.plot(lin['Time']*dt, lin['coords_a (2) (stats)']*1000, 
        lw=2,
        ls='--',
        color='k',
        label='Linear UVLM')

plt.plot(nlin['Time']*dt, nlin['coords_a (2) (stats)']*1000, 
        lw=2,
        ls='-',
        color='k',
        label='Nonlinear UVLM')

plt.xlabel('Time, t [s]')
plt.ylabel('Tip vertical displacement [mm]')
plt.title('Straight, u = 50 m/s, gust = 0.1%')
plt.grid()
plt.legend()
```

```python
lin = pd.read_csv('goland_swept_lin.csv')
nlin = pd.read_csv('goland_swept_nlin.csv')

dt = main_chord / 6/ 50
```

```python
fig = plt.figure(figsize=(12,10))

plt.plot(lin['Time']*dt, lin['coords_a (2) (stats)']*1000, 
        lw=2,
        ls='--',
        color='k',
        label='Linear UVLM')

plt.plot(nlin['Time']*dt, nlin['coords_a (2) (stats)']*1000, 
        lw=2,
        ls='-',
        color='k',
        label='Nonlinear UVLM')

plt.xlabel('Time, t [s]')
plt.ylabel('Tip vertical displacement [mm]')
plt.title('Sweep = 20deg, u = 50 m/s, gust = 0.1%')
plt.grid()
plt.legend()
plt.rcParams.update({'font.size':16})
```

```python
fig = plt.figure(figsize=(12,10))

plt.plot(lin['Time']*dt, lin['app_forces (2) (stats)'], 
        lw=2,
        ls='--',
        color='k',
        label='Linear UVLM')

plt.plot(nlin['Time']*dt, nlin['app_forces (2) (stats)'], 
        lw=2,
        ls='-',
        color='k',
        label='Nonlinear UVLM')

plt.xlabel('Time, t [s]')
plt.ylabel('Tip applied vertical force [N]')
plt.title('Sweep = 20deg, u = 50 m/s, gust = 0.1%')
plt.grid()
plt.legend()
plt.rcParams.update({'font.size':16})
```
